```python
breast_cancer.drop('Unnamed: 32', axis = 1, inplace=True)
```

```python
breast_cancer.drop('id', axis = 1, inplace=True)
```

```python
breast_cancer.columns.values
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
breast_cancer['diagnosis'] = le.fit_transform(breast_cancer['diagnosis'])

breast_cancer['diagnosis'].value_counts()
```

```python
breast_cancer.head()
```

```python
breast_cancer.info()
```

```python
data_vars = breast_cancer.columns.values.tolist()
# Label
y = ['diagnosis']
# features
X = [i for i in data_vars if i not in y]
```

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

**1. Random Forest**

```python
X_train, X_test, y_train, y_test = train_test_split(breast_cancer[X], breast_cancer[y], test_size = 0.15, random_state = 16)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

X_train = mm.fit_transform(X_train)
X_test = mm.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 50, max_depth = 25)

rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print("Training accuracy :", rf.score(X_train, y_train))
print("Testing accuarcy :", rf.score(X_test, y_test))
ranfor = rf.score(X_test, y_test)
randor_f1score = f1_score(y_test, y_pred)

cr_rf = classification_report(y_test, y_pred)
print(cr_rf)

cm1 = confusion_matrix(y_test, y_pred)
print(cm1)
plt.rcParams['figure.figsize'] = (5, 5)
sns.heatmap(cm1, annot = True, cmap = 'winter')
plt.title('Confusion Matrix', fontsize = 20)
plt.show()
```

```python
X_ = np.array(X)
importance_score = rf.feature_importances_
indices = np.argsort(importance_score)

feature_importances = pd.Series(rf.feature_importances_, index = X)
feature_importances.nlargest(10).plot(kind='barh')
```

**2. Support Vector Classifier**

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(breast_cancer[X], breast_cancer[y], test_size = 0.15, random_state = 16)

mm = MinMaxScaler()

X_train2 = mm.fit_transform(X_train2)
X_test2 = mm.fit_transform(X_test2)
```

```python
from sklearn.svm import SVC

model_svc = SVC()

model_svc.fit(X_train2, y_train2)
y_pred2 = model_svc.predict(X_test2)

print("Accuracy Training : ", model_svc.score(X_train2, y_train2))
print("Accuracy Testing : ", model_svc.score(X_test2, y_test2))
svc_f1score = f1_score(y_test2, y_pred2)

svc1 = model_svc.score(X_test2, y_test2)
cr_svr = classification_report(y_test2, y_pred2)
print(cr_svr)

cm2 = confusion_matrix(y_test2, y_pred2)
print(cm2)
plt.rcParams['figure.figsize'] = (5, 5)
sns.heatmap(cm2, annot = True, cmap = 'winter')
plt.title('Confusion Matrix', fontsize = 20)
plt.show()
```

**3. Logistic Regression**

```python
X_train3, X_test3, y_train3, y_test3 = train_test_split(breast_cancer[X], breast_cancer[y], test_size = 0.15, random_state = 16)

mm = MinMaxScaler()

X_train3 = mm.fit_transform(X_train3)
X_test3 = mm.fit_transform(X_test3)
```

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression()

model_lr.fit(X_train3, y_train3)

y_pred3 = model_lr.predict(X_test3)

print("Accuracy Training : ", model_lr.score(X_train3, y_train3))
print("Accuracy Testing : ", model_lr.score(X_test3, y_test3))
logreg_f1score = f1_score(y_test3, y_pred3)

lr1 = model_lr.score(X_test3, y_test3)
cr_lr = classification_report(y_test3, y_pred3)
print(cr_lr)

cm3 = confusion_matrix(y_test3, y_pred3)
print(cm3)
plt.rcParams['figure.figsize'] = (5, 5)
sns.heatmap(cm3, annot = True, cmap = 'winter')
plt.title('Confusion Matrix', fontsize = 20)
plt.show()
```

```python
the_model = ['SVC', 'Logistic Regression', 'Random Forest']
the_score = [svc1 ,lr1, ranfor]
the_f1_score = [svc_f1score, logreg_f1score, randor_f1score]

the_algoritma = {
    "Model Algoritma" : the_model,
    "Accuracy Testing" : the_score,
    "F1 Score" : the_f1_score
}

df_acc_score = pd.DataFrame(the_algoritma)
df_acc_score
```
